# Scrapping Cricket Rankings

## ICC Men's ODI Team Rankings

```python
page=requests.get('https://www.icc-cricket.com/rankings/mens/team-rankings/odi')
page
```

```python
MODI=BeautifulSoup(page.content)
Mteam=[]
Mmatches=[]
Mpoints=[]
Mratings=[]
for i in MODI.find_all('span',class_="u-hide-phablet"):
    Mteam.append(i.text)
fMteam=Mteam[0]
teams=Mteam[1:10]
fmatch=MODI.find('td',class_="rankings-block__banner--matches")
fMmatch=fmatch.text
fpoint=MODI.find('td',class_="rankings-block__banner--points")
fMpoint=fpoint.text
frating=MODI.find('td',class_="rankings-block__banner--rating u-text-right")
fMrating=frating.text.split(' \n'[0])[28]
for j in MODI.find_all('td',class_="table-body__cell u-center-text"):
    Mmatches.append(j.text)
    res_matches = [Mmatches[k] for k in range(len(Mmatches)) if k % 2 == 0]
    res_points = [Mmatches[l] for l in range(len(Mmatches)) if l % 2 != 0]
rem_matches=res_matches[0:9]
rem_points=res_points[0:9]
for m in MODI.find_all('td',class_="table-body__cell u-text-right rating"):
    Mratings.append(m.text)
rem_ratings=Mratings[0:9]
dfw=pd.DataFrame({'MEN_TEAM':[fMteam],'MATCHES':[fMmatch],'POINTS':[fMpoint],'RATINGS':[fMrating]})
dfw1=pd.DataFrame({'MEN_TEAM':teams,'MATCHES':rem_matches,'POINTS':rem_points,'RATINGS':rem_ratings})
Final_Team=pd.concat([dfw,dfw1],axis=0)
Final_Team.index = np.arange(1, len(Final_Team) + 1)
df5a=pd.DataFrame(Final_Team)
print('Mens Team Rankings')
df5a
```

```python
ODIBat=BeautifulSoup(player.content)
ODIBat
```

```python
batsman=[]
team=[]
rating=[]
fbatsman=ODIBat.find('div',class_="rankings-block__banner--name")
nfbat=fbatsman.text
fteam=ODIBat.find('div', class_="rankings-block__banner--nationality")
nfteam=fteam.text.split('\n')[2]
frating=ODIBat.find('div',class_="rankings-block__banner--rating")
nfrating=frating.text
for i in ODIBat.find_all('td',class_="table-body__cell name"):
    batsman.append(i.text.split('\n')[1])
nbat=batsman[0:9]
for j in ODIBat.find_all('td',class_="table-body__cell nationality-logo"):
    team.append(j.text.split('\n')[2])
nteam=team[0:9]
for k in ODIBat.find_all('td',class_="table-body__cell u-text-right rating"):
    rating.append(k.text)
nrating=rating[0:9]
df51b=pd.DataFrame({'Batsman':nbat,'Team':nteam,'Rating':nrating})
df51=pd.DataFrame({'Batsman':[nfbat],'Team':[nfteam],'Rating':[nfrating]})
Final_Batsman=pd.concat([df51,df51b],axis=0)
Final_Batsman.index = np.arange(1, len(Final_Batsman) + 1)
df5b=pd.DataFrame(Final_Batsman)
print('Mens Top-10 ODI Batsman Rankings')
df5b
```

# Scrapping details from dineout

```python
dine=requests.get('https://www.dineout.co.in/hyderabad-restaurants/central-east-hyderabad/banjara-hills')
dine
```

```python
eat=BeautifulSoup(dine.content)
eat
```

```python
name=[]
cuisine=[]
location=[]
ratings=[]
image=[]
for i in eat.find_all('a',class_="restnt-name ellipsis"):
    name.append(i.text)
for j in eat.find_all('div',class_="restnt-loc ellipsis"):
    location.append(j.text.split(',')[0])
for k in eat.find_all('div',class_="restnt-rating rating-4 hide"):
    ratings.append(k.text)
for m in eat.find_all('img',class_="no-img"):
    image.append(m.get('data-src'))
for l in eat.find_all('span',class_="double-line-ellipsis"):
    cuisine.append(l.text.split('|')[1])
import pandas as pd
df9=pd.DataFrame({'Restaurant_Name':name,'Cuisine':cuisine,'Location':location,'Ratings':ratings,'Image_URL':image})
df9
```
