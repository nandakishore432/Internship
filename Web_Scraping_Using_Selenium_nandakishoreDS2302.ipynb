# Q4. Scrape data of first 100 sunglasses listings on flipkart.com

```python
#Lets first connect to the driver and open chrome
driver=webdriver.Chrome(r"C:\Users\Vipinchi Nidhi\Downloads\chromedriver.exe")
```

```python
#Openng flipkart.com page on automated chrome browser to scrape details
driver.get('https://www.flipkart.com/')
```

```python
#Let's Import required selenium and pandas 
import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
from selenium.webdriver.common.by import By
import time
```

```python
sunglasses=driver.find_element(By.CLASS_NAME,"_3704LK")
sunglasses.send_keys('sunglasses')
```

```python
search=driver.find_element(By.CLASS_NAME,"L0Z3Pu")
search.click()
```

```python
sg_name=[]
sg_description=[]
sg_price=[]
sg_discount=[]
```

```python
#Scrapping 100 Sunglasses from next papges
start=0
end=2
for page in range(start,end):
    names=driver.find_elements(By.XPATH,'//div[@class="_2WkVRV"]')
    descriptions=driver.find_elements(By.XPATH,'//a[@class="IRpwTa"]')
    prices=driver.find_elements(By.XPATH,'//div[@class="_30jeq3"]')
    discounts=driver.find_elements(By.XPATH,'//div[@class="_3Ay6Sb"]')
    for i in names:
        sg_name.append(i.text)
    for j in descriptions:
        sg_description.append(j.text)
    for k in prices:
        sg_price.append(k.text)
    for l in discounts:
        sg_discount.append(l.text)
    next_button=driver.find_element(By.XPATH,'//a[@class="_1LKTO3"]')
    next_button.click()
    time.sleep(3)
tnames=sg_name[0:100]
tdes=sg_description[0:100]
tprice=sg_price[0:100]
tdisc=sg_discount[0:100]
```

```python
import pandas as pd
df4=pd.DataFrame({'Brand Name':tnames,'Information':tdes,'Cost':tprice,'Offers':tdisc})
df4                                                        
```

```python
pm_name=[]
pm_born_dead=[]
pm_term=[]
pm_remarks=[]
```
