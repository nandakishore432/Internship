# Q4. Scrape the details of trending repositories on Github.com.

```python
#lets connect with the webdriver
driver=webdriver.Chrome(r"C:\Users\Vipinchi Nidhi\Downloads\chromedriver.exe")
```

```python
#Lets Open the Github.com web page from Google Chrome
driver.get("https://github.com/")
```

```python
#Let's find Open source or Explore Options to find Trending Repositories
opn_src=driver.find_element(By.XPATH,'/html/body/div[1]/div[1]/header/div/div[2]/div/nav/ul/li[3]/button')
opn_src.click()
```

```python
#Now Open the Trending Repositories by Code
trend=driver.find_element(By.XPATH,'/html/body/div[1]/div[1]/header/div/div[2]/div/nav/ul/li[3]/div/div[3]/ul/li[2]/a')
trend.click()
```

```python
# Scraping the Name of the repositories
Name=[]
name=driver.find_elements(By.XPATH,'//span[@class="text-normal"]')
for i in name:
    try:
        Name.append(i.text)
    except NoSuchElemenetException:
        Name.append("--")
print(len(Name),Name)
```

```python
#Check the lentgh of the Repositories names
len(Name)
```

```python
#Scrap the URLs of the repositories for scraping exclusive details
rep_url=[]
url=driver.find_elements(By.XPATH,'//h2[@class="h3 lh-condensed"]/a')
for i in url:
    rep_url.append(i.get_attribute("href"))
print(len(rep_url),rep_url)
```

```python
#Scraping the description of the repositories
Desc=[]
desc=driver.find_elements(By.XPATH,'//p[@class="col-9 color-fg-muted my-1 pr-4"]')
for i in desc:
    try:
        Desc.append(i.text)
    except NoSuchElementException:
        Desc.append("__")
print(len(Desc),Desc)
```

```python
#Scrap the number of the contributors from each URL individually
Contributors=[]
for i in rep_url:
    driver.get(i)     #Obtain the URL from rep_url
    time.sleep(2)
    cont=driver.find_elements(By.XPATH,'//*[@id="repo-content-pjax-container"]/div/div/div[2]/div[2]/div/div[3]/div/h2/a/span')
    for i in cont:
        try:
            Contributors.append(i.text)   #append the data of contributors number
        except NoSuchElemenetException:
            Contributors.append("__")
print(len(Contributors),Contributors) #Check the length and print the number of contributors
```

```python
#Scrap the Ratings of the Repositories
Ratings=[]
rating=driver.find_elements(By.XPATH,'//span[@ class="d-inline-block float-sm-right"]')
for i in rating:
    try:
        Ratings.append(i.text.replace("stars today",""))
    except NoSuchElemenetException:
        Ratings.append("__")
print(len(Ratings),Ratings)
```

```python
#Scraping the number of the languages used and implemented in the Repository
Languages=[]
lang=driver.find_elements(By.XPATH,'//span[@itemprop="programmingLanguage"]')
for i in lang:
    if i.text is None:
        Languages.append("__")
    else:
        Languages.append(i.text)
print(len(Languages),Languages)
```

```python
#Make the scraped details into a list for easy DataFrames
data=list(zip(Name,rep_url,Desc,Languages,Ratings))
#Create a DataFrame of the details scraped and display them
DF4=pd.DataFrame(data,columns=["Repository Name","Repository URL","Repository Description","Languages Used","Star Count"])
DF4
```
